```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/proxy_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
signal_cols = ['availability_score','latency_score','anonymity_score','protocol_support_score','reputation_score','duplicate_risk_score']
signal_avgs = df[signal_cols].mean()
colors = ['#14b8a6','#3b82f6','#f97316','#8b5cf6','#10b981','#ef4444']
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('BHProxy - Average Proxy Signal Scores', fontsize=14, fontweight='bold')
plt.xlabel('Proxy Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=60, color='green', linestyle='--', alpha=0.5, label='Healthy threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
anon_counts = df['anonymity_level'].value_counts()
plt.figure(figsize=(8,5))
plt.pie(anon_counts, labels=anon_counts.index, colors=['#10b981','#f59e0b','#ef4444'], autopct='%1.0f%%')
plt.title('BHProxy - Anonymity Level Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
